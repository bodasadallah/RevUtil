```python
import pandas as pd
import datasets
```

```python
df = pd.read_csv('/home/abdelrahman.sadallah/mbzuai/review_rewrite/data/1k_samples_human_annotation_sampled.csv')

```

```python
df
```

```python
## reomvoe rows that has the value 1 in column E
df = df[df['Discard'] != 1.0]
```

```python
df = df[df['Include in 25']== 1]
```

```python
len(df)
```

```python
## shuffle the rows
df = df.sample(frac=1).reset_index(drop=True)

# remove column 'Discard'
df = df.drop(columns=['Discard', 'Include in 25'])
df
```

```python
## split this df into 5 dataframes with sizes 100, 160, 200, 200, 200
df_100 = df[:100]
df_160 = df[100:260]
df_200 = df[260:460]
df_200_2 = df[460:660]
df_200_3 = df[660:860]
## convert df to hf datasets
df_100 = datasets.Dataset.from_pandas(df_100)
df_160 = datasets.Dataset.from_pandas(df_160)
df_200 = datasets.Dataset.from_pandas(df_200)
df_200_2 = datasets.Dataset.from_pandas(df_200_2)
df_200_3 = datasets.Dataset.from_pandas(df_200_3)
```

```python
## load HF authentication token from .env file
import os
from dotenv import load_dotenv
load_dotenv()
hf_token = os.getenv("HF_TOKEN")


## push the datasets to HF hub

df_100.push_to_hub("boda/review_evaluation_main_1",token=hf_token)
df_160.push_to_hub("boda/review_evaluation_main_2",token=hf_token)
df_200.push_to_hub("boda/review_evaluation_main_3",token=hf_token)
df_200_2.push_to_hub("boda/review_evaluation_main_4",token=hf_token)
df_200_3.push_to_hub("boda/review_evaluation_main_5",token=hf_token)
```
