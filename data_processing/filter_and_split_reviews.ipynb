```python
import os
import sys
from pathlib import Path
import pandas as pd
from tqdm import tqdm
import re
import importlib


module_path_mine = Path(os.path.abspath(""))
module_path = Path(os.path.abspath("")).parent
print(module_path)
sys.path.append(str(module_path))
sys.path.append(str(module_path_mine))
sys.path.append(str(module_path)+ "/inference")



from inference.utils import *
from inference.rule_based_review_split import *

# importlib.reload(utils)


processed_folder_path = f'{module_path}/data/processed'
```

```python
confs = {}
files = os.listdir(processed_folder_path)

# Step 2: Concat reviews from all sources
for file in files:
    file_path = os.path.join(processed_folder_path, file)
    if os.path.isfile(file_path) and file.endswith('.csv'):
        print(f'Processing file: {file}')
        name = file.split('.')[0]
        if 'iclr24' in file:
            confs [name] = pd.read_csv(file_path, dtype=str, escapechar='\\')
        else:
            confs [name] =pd.read_csv(file_path, dtype=str)

```

```python
all_reviews = []
```

```python
############################ Process ACL reviews ############################
for i,paper in confs['acl2017_reviews'].iterrows():
    venue = '_'.join(paper['filename'].split('_')[:2])
    ## Remove Weaknesses from focused review if it's in the beginning\
    focused_review = paper['focused_review']
    paper_id = paper['filename'].split('.json')[0]
    all_reviews.append({'paper_id': paper_id, 'venue': venue, 'focused_review': focused_review})
```

```python
############################ Process ARR reviews ############################
for i,paper in confs['reviewer2_ARR_2022_reviews'].iterrows():
    venue = '_'.join(paper['paped_id'].split('_')[:2])
    ## Remove Weaknesses from focused review if it's in the beginning\
    focused_review = paper['focused_review']
    paper_id = paper['paped_id'].split('.json')[0]
    all_reviews.append({'paper_id': paper_id, 'venue': venue, 'focused_review': focused_review})
```

```python
############################ Process old ICLR reviews ############################
for i,paper in confs['iclr_reviews'].dropna(subset=['focused_review']).iterrows():
    venue = '_'.join(paper['review_id'].split('_')[:2])
    ## Remove Weaknesses from focused review if it's in the beginning\
    focused_review = paper['focused_review']
    paper_id = paper['review_id'].split('.json')[0]
    all_reviews.append({'paper_id': paper_id, 'venue': venue, 'focused_review': focused_review})
```

```python
############################ Process NIPS reviews ############################
for i,paper in confs['nips_reviews'].dropna(subset=['focused_review']).iterrows():
    venue = '_'.join(paper['review_id'].split('_')[:2])
    ## Remove Weaknesses from focused review if it's in the beginning\
    focused_review = paper['focused_review']
    paper_id = paper['review_id'].split('.json')[0]
    all_reviews.append({'paper_id': paper_id, 'venue': venue, 'focused_review': focused_review})
```

```python
############################ Process ICLR 2024,2025 and EMNLP 2023 reviews ############################
for i,paper in confs['iclr24,25_emnlp23'].dropna(subset=['focused_review']).iterrows():
    venue = '_'.join(paper['venue'].split(' '))
    ## Remove Weaknesses from focused review if it's in the beginning\
    focused_review = paper['focused_review']
    paper_id = paper['arr_id']
    all_reviews.append({'paper_id': paper_id, 'venue': venue, 'focused_review': focused_review})
```

```python
import ast


all  = pd.read_csv(f'{module_path}/data/all_reviews.csv')

num_of_points = []
no_points = 0

sources_cnt = {}
sources_points_stats = {}
for i,r in all.iterrows():
    points = ast.literal_eval(r['split_review'])
    src = r['venue']
    if len(points) == 1:
        no_points += 1
    else:
        sources_points_stats[src] = sources_points_stats[src] + len(points) if src in sources_points_stats else len(points)
    sources_cnt[src] = sources_cnt[src] + 1 if src in sources_cnt else 1
    num_of_points.append(len(points))


print(f'Number of reviews: {len(all)}')
print(f'Number of reviews with one point: {no_points}')
print(f'Number of reviews with more than one point: {len(all) - no_points}')
print(f'Precentage of reviews with one point: {no_points/len(all)}')
print(f'total number of points: {sum(num_of_points)}')
print(f'Average number of points: {sum(num_of_points)/len(num_of_points)}')
print(f'Max number of points: {max(num_of_points)}')
print(f'Min number of points: {min(num_of_points)}')

    


print(f'Number of samples of each source with more that one point')
for source, count in sources_cnt.items():
    print(f'{source}: {count} ')

print(f'Number of points for each source')
for source, count in sources_points_stats.items():
    print(f'{source}: {count}')
```
